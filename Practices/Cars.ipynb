## Data Cleaning 

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

## Data Transformation 

```python
df.rename(columns = {'Make' : 'Name'}, inplace = True)
```

```python
df.head(1)
```

```python
df['Horsepower'] = df['Horsepower'].astype('int64')
```

```python
df.info()
```

## Data Filtering and Subsetting 

```python
filtered_df = df[df['Horsepower'] > 400 ]
```

```python
filtered_df
```

```python
subset = df[['Name','Origin']]
subset = subset.drop_duplicates()
```

```python
subset
```

## Group By and Aggregation

```python
grouped = df.groupby('Type').count()
```

```python
grouped
```

```python
grouped['MPG_City'].mean()
```
